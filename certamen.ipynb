### Ejercicio 3.1 – Calculadora de propina (12 pts)

Escribe un programa que:
1. Pida el **total de la cuenta** (puede ser decimal).
2. Pida el **porcentaje de propina** (número entero).
3. Calcule la **propina** y el **total a pagar**.

**Salida esperada:**
```
Ingresa el total de la cuenta: 8500
Ingresa el porcentaje de propina (%): 10
----------------------------
Propina: $850.0
Total a pagar: $9350.0
```



````python
# Ejercicio 3.1 – Calculadora de propina
# Escribe tu código aquí:
total = float(input("Ingrese el total de la cuenta: "))
print("total de la cuenta: ", total)
prop = float(input("ingrese propina (10% minimo)"))
print("total de la propina: ", prop)
print("Propina: ", (prop/100) * total)
print("Total a pagar: ", total + ((prop/100) * total))
````

### Ejercicio 3.2 – Clasificador de notas (16 pts)

Escribe un programa que:
1. Pida una **nota entre 1.0 y 7.0**.
2. Valide el rango → si no es válida, muestra `"Nota inválida."` y termina.
3. Muestre el **concepto** y si está **aprobado o reprobado**.

| Rango        | Concepto      |
|--------------|---------------|
| 6.0 – 7.0    | Excelente     |
| 5.0 – 5.9    | Bueno         |
| 4.0 – 4.9    | Suficiente    |
| 1.0 – 3.9    | Insuficiente  |

**Salida esperada:**
```
Ingresa tu nota (1.0 - 7.0): 5.5
Concepto: Bueno
Estado: Aprobado
```


````python
# Ejercicio 3.2 – Clasificador de notas
# Escribe tu código aquí:
cali = float(input("Ingresa tu nota"))
print("Ingresa tu nota (1.0 - 7.0): ", cali)
if cali >= 6.0 and cali <= 7.0:
    print("Concepto: Excelente")
    print("Estado: Aprobado")
elif cali >= 5.0 and cali <= 5.9:
    print("Concepto: Bueno")
    print("Estado: Aprobado")
elif cali >= 4.0 and cali <= 4.9:
    print("Concepto: Suficiente")
    print("Estado: Aprobado")
elif cali >= 1.0 and cali <= 3.9:
    print("Concepto: Insuficiente")
    print("Estado: Reprobado")
else:
    print("Datos incorrectos")
````

### Ejercicio 3.3 – Gestión de lista de compras (16 pts)

Usando la lista inicial `["leche", "pan", "huevos"]`:
1. Muestra la lista inicial.
2. Agrega `"mantequilla"` y `"jugo"`.
3. Elimina `"pan"`.
4. Muestra cuántos productos hay.
5. Recorre con `for` y muestra cada producto numerado desde 1.

**Salida esperada:**
```
Lista inicial: ['leche', 'pan', 'huevos']
Lista actualizada: ['leche', 'huevos', 'mantequilla', 'jugo']
Total de productos: 4
1. leche
2. huevos
3. mantequilla
4. jugo
```


````python
# Ejercicio 3.3 – Gestión de lista de compras
# Escribe tu código aquí:
listas = ["leche", "pan", "huevos"]
lista_inicial = listas
print("Lista inicial: ",lista_inicial)
listas.append("mantequilla")
listas.append("jugo")
listas.remove("pan")
print("Lista actualizada: ", lista_inicial)
print("Total de productos: ", len(listas))
for i in listas:
    print(i)
````

### Ejercicio 3.4 – Adivina el número (16 pts)

- El número secreto es `secreto = 42` (fijo, **no lo pidas al usuario**).
- Usa `while` para pedir intentos repetidamente.
- En cada intento indica si el número es **mayor** o **menor**.
- Al adivinar, muestra cuántos intentos necesitó.

**Salida esperada:**
```
Adivina el número secreto (entre 1 y 100):
Intento 1: 50
Es menor. Intenta de nuevo.
Intento 2: 30
Es mayor. Intenta de nuevo.
Intento 3: 42
¡Correcto! Adivinaste en 3 intentos.
```


````python
# Ejercicio 3.4 – Adivina el número
# Escribe tu código aquí:
print("Adivina el numero secreto (entre 1 y 100): ")
num_sec = 42
intento = int(input("ingrese un numero del 1 al 100"))
while True:
    if intento < num_sec:
        print(("Es menor. Intenta de nuevo"))
    elif intento > num_sec:
        print(("Es Mayor. Intenta de nuevo"))
    else:
        print("ganaste")
        break


````
